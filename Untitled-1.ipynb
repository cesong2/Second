### 소비자물가지수 증가에따른 건물거래금액변화/ 물가지수 상승 = 건물가 상승
rvalue = 0.3 pvalue = 0

```python
x=df0['소비자물가지수']
x=list(x)
y=df0['거래금액']
y=list(y)
```

```python
result = stats.linregress(x,y)
result
```

```python
slope, intercept, r_value, p_value, stderr = stats.linregress(x, y)
x1 = np.array(x)
#산점도 그리기
plt.scatter(x,y)
#회귀선 그리기
plt.plot(x1, slope*x1 +intercept, c="red")
plt.xlabel("소비자물가지수")
plt.ylabel("건물거래금액")
```
